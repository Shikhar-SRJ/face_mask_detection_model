```python
from keras.models import load_model
import cv2
import numpy as np
```

```python
model= load_model('/home/suraj/tensorflow/Notebooks/Face Mask/Face mask detector/model_last.h5')

face_clsfr = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')


# label_dict = {0:'MASK', 1:"NO MASK"}
# color_dict = {0:(0,255,0), 1:(0,0,255)}

label_dict = {0:"NO MASK", 1:'MASK'}
color_dict = {0:(0,0,255), 1:(0,255,0)}
```

```python
model.summary()
```

```python
source = cv2.VideoCapture(0)

img_size=200

while True:
    ret, img = source.read()
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = face_clsfr.detectMultiScale(gray, 1.3, 5)
    
    for x, y, w ,h in faces:
        face_img = img[y:y+h, x:x+w]
        resized = cv2.resize(face_img, (img_size,img_size))
        normalized = resized/255.0
        reshaped = np.reshape(normalized, (1,img_size,img_size, 3))
        result = model.predict(reshaped)
        
        label = np.argmax(result, axis=1)[0]
        
        cv2.rectangle(img,(x,y),(x+w,y+h), color_dict[label], 2)
        cv2.rectangle(img,(x,y-4),(x+w,y), color_dict[label], -1)
        cv2.putText(img, label_dict[label], (x,y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255,255,255), 2)
        cv2.putText(img, str(np.max(result)), (x+130,y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255,255,255), 2)
        
    
    cv2.imshow("LIVE", img)
    key = cv2.waitKey(1)
    
    if (key==27):
        break
    
cv2.destroyAllWindows()
source.release()
```
